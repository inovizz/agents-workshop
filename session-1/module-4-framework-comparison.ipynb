## Why We Chose LangGraph

For this workshop, we're using **LangGraph** because:

### 1. Graph-Based Control Flow
- Visual understanding of agent logic
- Explicit state management
- Support for cycles (essential for ReAct)

### 2. Production Features
- Built-in checkpointing/persistence
- Human-in-the-loop patterns
- Streaming support
- Observability with LangSmith

### 3. Flexibility
- Works with any LLM (OpenAI, Anthropic, local)
- Custom agent architectures
- Low-level control when needed

### 4. Active Development
- LangChain 1.0 integration
- Regular updates
- Strong community

````python
import sys
sys.path.insert(0, '..')  # Add parent directory to path for setup_llm

from setup_llm import get_chat_model  # Use LLM helper
from langchain_core.tools import tool
from langchain.agents import create_agent  # LangChain v1.0 - new standard for building agents

# Define tool
@tool
def get_weather(city: str) -> str:
    """Get weather for a city."""
    return f"Weather in {city}: 22Â°C, Sunny"

# Initialize model using DIAL (Azure OpenAI)
model = get_chat_model(temperature=0)

# Create agent using create_agent (LangChain v1.0)
agent = create_agent(
    model=model,
    tools=[get_weather],
    system_prompt="You are a helpful weather assistant."
)

# Run
result = agent.invoke({"messages": [("user", "What's the weather in Tokyo?")]})
print(f"âœ… LangGraph Agent Result: {result['messages'][-1].content}")
````
